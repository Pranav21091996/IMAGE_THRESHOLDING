```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im = cv2.imread("threshold.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(im)
```

```python
thresh = 0
maxValue = 255
_, dst = cv2.threshold(im, thresh, maxValue, cv2.THRESH_BINARY)
plt.imshow(dst)
```

```python
_, dst = cv2.threshold(im, thresh, maxValue, cv2.THRESH_BINARY_INV)
plt.imshow(dst)
```
